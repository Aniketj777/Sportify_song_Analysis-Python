#**Exploratory Data Analysis (EDA)**

####**1). Top 5 artist popular artist**

```python
top_5_artist = df.groupby("artist").count().sort_values(by = "song_title", ascending=False)['song_title'][:5]
top_5_artist
```

```python
top_5_artist.plot.barh()
plt.show()
```

So from the above bar chart we can say that DRAKE has rated most popular artist
and WALK THE MOON at 5 most top artist.

####**2).Top 5 loudest track**

```python
top_five_loudest = df[['loudness','song_title']].sort_values("loudness", ascending=True)[:5]
top_five_loudest
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x = "loudness", y = "song_title", data = top_five_loudest)
plt.title("Top Five Loudest Track")
plt.show()
```

From the above plot we can say that "Lyric Pieces, Book I Op. 12: I. Arietta" is the loudest track where "8 Fantasiestcke, Op.12 : 1. Des Abends" is the second loudest track.
And "Eleanor" is the fifth loudest track.

####**3).Artist with most danceability song**

```python
top_artiest_danceability = df[['danceability', 'song_title', "artist"]].sort_values(by = "danceability", ascending = False)[:5]
top_artiest_danceability
```

```python
plt.figure(figsize=(13,7))
sns.barplot(x = "danceability", y="artist", data = top_artiest_danceability)
```

From the above bar plot we can say that "Ben Remember" is the artist which has highest danceability tracks.

Where as "Justin Timberlake" is second highest rated artist for danceability.

####**4).Top Instrumental tracks**

```python
top_instrumental_tracks = df[['instrumentalness','song_title', 'artist']].sort_values(by ='instrumentalness', ascending=False)[:5]
top_instrumental_tracks
```

```python
plt.figure(figsize = (12,7))
plt.pie(x=top_instrumental_tracks["instrumentalness"], data=top_instrumental_tracks, autopct = "%1.2f%%",labels=top_instrumental_tracks['song_title'])
plt.show()
```

So from the above pie chart we get that top instrumental track are "Senseless Order"

where as second highest top instrument tracks are "So Destroyed (channeling Rage Peace)	"

And at fifth place "Strandbar (disko)" is the top 5th instrumental track.

####**5).Multiple Feature plots**

```python
feature_columns = ["tempo","loudness","acousticness","danceability","duration_ms","energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for i in feature_columns:
  pos_data = df[df['target'] == 1][i]
  neg_data = df[df['target'] == 0][i]

  plt.figure(figsize=(10,5))
  sns.distplot(pos_data, bins=30, label='Positive', color='Green')
  sns.distplot(neg_data, bins=30, label='Negative', color='Red')
  plt.legend(loc='best')
  plt.title(f"Positive & Negative histogram plot for{i}")
  plt.show()
```

In the above charts Red bar shows negative data and green shows positive
